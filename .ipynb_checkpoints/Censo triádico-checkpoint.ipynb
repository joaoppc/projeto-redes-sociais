```python
def ranking(gml,string):
    lista=[]
    for i in range(20):
        lista.append((gml.nodes[i]['label'],float(gml.nodes[i][string])))
    return sorted(lista, key=lambda x: float(x[1]), reverse = True)
    
```

```python
a2018.set_all_nodes('size', 5)
a2018.set_all_nodes('labpos', 'hover')
a2018.move('kamada_kawai')
a2018.draw()
a2018.interact()
```

```python
tuplas_teams=[]
count = 0
lista_regress = []
passed1 = False
dict_teams={}
valor_regress = 0
for i in range(20):
    for j in range(20):
        for h in range(20):
            if j == i or j == h or i == h:
                pass
            else:
                tuplas_teams.append((a2018.nodes[i],a2018.nodes[j]))
                try:
                    passed1 = False
                    a2018.edges[h,i]#c ganhou de a
                    lista_regress.append(int(-1))
                    valor_regress = 0
                    passed1 = True
                    a2018.edges[i,h]#a ganhou de c
                    lista_regress.append(int(1))
                    valor_regress = 1
                except:
                    if passed1 == False:
                        lista_regress.append(int(0))
                        valor_regress = 0
                try:
                    print(a2018.nodes[i]['label'],a2018.nodes[j]['label'],a2018.nodes[h]['label'])
                    print('a - b',a2018.edges[i,j])
                    print('b - c',a2018.edges[j,h])
                    count += 1
                    print('contador',count)
                except:
                    pass
            dict_teams[a2018.nodes[i]['label']+a2018.nodes[h]['label']]=[int(a2018.nodes[i]['position']),int(a2018.nodes[h]['position']),count, valor_regress,int(a2018.nodes[i]['position'])-int(a2018.nodes[h]['position'])]
        count = 0
#print(dict_teams)
        
print(dict_teams['VitoriaSao Paulo'])
```

```python
import csv
row = ['ranking_a','ranking_c','simmelianos','resultado','diferenca_rk']


with open('regress.csv', mode = 'w') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(row)
    for x in dict_teams.values():
        
        writer.writerow(x)
        
csv_file.close()
```
