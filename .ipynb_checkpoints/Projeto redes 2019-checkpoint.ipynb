```python
import networkx as nx
import freeman as fm
import pandas as pd
```

```python

a2019 = fm.load('dados/result_teams_brA2019.gml')
#a2018 = fm.load('dados/result_teams_brA2018.gml')
g = nx.read_gml('dados/result_teams_brA2018.gml', 'id')
a2017 = fm.load('dados/result_teams_brA2017.gml')
a2016 = fm.load('dados/result_teams_brA2016.gml')

#b2019 = fm.load('dados/result_teams_brB2019.gml')
b2018 = fm.load('dados/result_teams_brB2018.gml')
b2017 = fm.load('dados/result_teams_brB2018.gml')
b2016 = fm.load('dados/result_teams_brB2018.gml')


```

```python
g.edges
```

```python
nx.draw(g)
```

```python
a2018.set_all_nodes('size', 5)
a2018.set_all_nodes('labpos', 'hover')
a2018.move('kamada_kawai')
a2018.draw()
a2018.interact()
```

```python
tuplas_teams=[]
count = 0
lista_regress = []
passed1 = False
dict_teams={}
valor_regress = 0
for i in range(20):
    for j in range(20):
        for h in range(20):
            if j == i or j == h or i == h:
                pass
            else:
                tuplas_teams.append((g.nodes[i],g.nodes[j]))
                try:
                    passed1 = False
                    a2018.edges[h,i]#c ganhou de a
                    lista_regress.append(int(-1))
                    valor_regress = 0
                    passed1 = True
                    a2018.edges[i,h]#a ganhou de c
                    lista_regress.append(int(1))
                    valor_regress = 1
                except:
                    if passed1 == False:
                        lista_regress.append(int(0))
                        valor_regress = 0
                for t in range(2):
                    try:
                        print(a2018.nodes[i]['label'],a2018.nodes[j]['label'],a2018.nodes[h]['label'])
                        print('a - b',a2018.edges[i,j,t])
                        print('b - c',a2018.edges[j,h,t])
                        count += 1
                        print('contador',count)
                    except:
                        pass
                
            dict_teams[a2018.nodes[i]['label']+a2018.nodes[h]['label']]=[int(a2018.nodes[i]['position']),int(a2018.nodes[h]['position']),count, valor_regress,int(a2018.nodes[i]['position'])-int(a2018.nodes[h]['position'])]
        count = 0
```

```python
import csv
row = ['ranking_a','ranking_c','simmelianos','resultado','diferenca_rk']


with open('regress.csv', mode = 'w') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(row)
    for x in dict_teams.values():
        
        writer.writerow(x)
        
csv_file.close()
```

```python
regress = pd.read_csv('regress.csv')
fm.logregress(regress,['diferenca_rk','simmelianos'],'resultado')
```
